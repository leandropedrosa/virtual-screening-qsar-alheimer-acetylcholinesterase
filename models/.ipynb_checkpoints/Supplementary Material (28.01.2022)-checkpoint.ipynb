**Preprocessing Step 1**

Model development using: PCA

Applying dimensionality reduction using PCA

StandardScaler to transform the X_scaled value

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
```

**Preprocessing Step 2**

Applying PCA

```python
from sklearn.decomposition import PCA
pca_20 = PCA(n_components=20,random_state=2)
pca_20.fit(X_scaled)
X_pca_20 = pca_20.transform(X_scaled)
```

**Checking the Explained variance ratio**

i.e. to know about how well the PCA components are represented without any loss between the variables (selected descriptors).

In our case it captured 99.98%. i.e. the variables (descriptor informations) are not missed their information and presented 99.98% within this 20 selected Principle components.

```python
print('Variance explained by all 20 principle components =' , sum(pca_20.explained_variance_ratio_ *100))
```

```python
np.cumsum(pca_20.explained_variance_ratio_ *100)
```

```python
plt.plot(np.cumsum(pca_20.explained_variance_ratio_))
plt.xlabel('No. of Components')
plt.ylabel('Explained variance')
plt.savefig('Elbow_plot.png',dpi=100)
```

```python
print('Variance explained by 1st PC =', np.cumsum(pca_20.explained_variance_ratio_ *100)[0])
print('Variance explained by 2nd PC =', np.cumsum(pca_20.explained_variance_ratio_ *100)[1])
print('Variance explained by 3rd PC =', np.cumsum(pca_20.explained_variance_ratio_ *100)[2])
print('Variance explained by 4th PC =', np.cumsum(pca_20.explained_variance_ratio_ *100)[3])
print('Variance explained by 5th PC =', np.cumsum(pca_20.explained_variance_ratio_ *100)[4])
print('Variance explained by 6th PC =', np.cumsum(pca_20.explained_variance_ratio_ *100)[5])
print('Variance explained by 7th PC =', np.cumsum(pca_20.explained_variance_ratio_ *100)[6])
```

**Applying PCA with n=2**

```python
pca_2 = PCA(n_components=2, random_state=2)
pca_2.fit(X_scaled)
X_pca_2 = pca_2.transform(X_scaled)
```

**2D Plot of 90% variability captured via PCA**

Plot for 2 Principle Components

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.scatterplot(x=X_pca_2[:,0], y=X_pca_2[:,1], s=70, hue=Y)
plt.title('2D Scatter Plot: 90% Variability Captured')
plt.xlabel('First Principle Component')
plt.ylabel('Second Principle Component')
plt.savefig('2D Scatter Plot')
```

**Applying PCA with n=3**

```python
pca_3 = PCA(n_components=3, random_state=2)
pca_3.fit(X_scaled)
X_pca_3 = pca_3.transform(X_scaled)
```

**Plot for 3 Principle Components**

```python
from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection='3d')
sctt = ax.scatter3D(X_pca_3[:,0], X_pca_3[:,1],X_pca_3[:,2], c = Y, s = 50, alpha = 0.6)
plt.title('3D Scatter Plot')
ax.set_xlabel('First Principle Component')
ax.set_ylabel('Second Principle Component')
ax.set_zlabel('Third Principle Component')
plt.savefig('3D Scatter Plot')
```

**Applying PCA setting n_components=0.90**

This will pick adequate principle components to preserve 90% variability

```python
pca_90 = PCA(n_components=0.90, random_state=1)
pca_90.fit(X_scaled)
X_pca_90 = pca_90.transform(X_scaled)
```

Picking adequate number of Principle Components to preserve 90% of variability

In our case it picked 5 Principle Components


```python
X_pca_90.shape
```

**Visualizing the Principle components**


```python
plt.figure(figsize=(10,7))
plt.plot(X_pca_90)
plt.xlabel('Observation')
plt.ylabel('Transformed')
plt.title('Transformed data by principle components (90% variabiity)')
plt.savefig('Principle components 90% Variability')
```

**Creating pandas DataFrame for the 5 principle components** 

So that it can be used for further regression analysis

```python
df_new = pd.DataFrame(X_pca_90,columns=['PC1','PC2','PC3','PC4','PC5'])
df_new['target'] = Y
df_new.head()
```
