# Exploração

A média, mediana e moda são medidas estatísticas que ajudam a resumir e entender uma base de dados de forma simples.

Média: é o valor obtido pela soma dos valores da base de dados dividido pelo número de valores. Ela mostra a tendência central dos dados, indicando qual é o valor médio da base de dados.

Mediana: é o valor que ocupa o lugar central dos dados quando eles estão ordenados. Ela é menos afetada por valores extremos do que a média.

Moda: é o valor mais frequente na base de dados. Ela indica qual é o valor mais comum na base de dados.

É importante notar que cada medida estatística tem suas aplicações e limitações e deve ser utilizada de acordo com a necessidade do problema. Por exemplo, se existirem muitos outliers (valores extremos) na base de dados, a mediana é uma medida mais indicada do que a média.

O Skewness (assinatura ou assimetria) é uma medida estatística que indica a assimetria de uma distribuição de dados em relação à sua média. Ele pode ser positivo, negativo ou zero.

Skewness positivo: indica que a distribuição tem uma cauda mais longa à direita (mais valores à direita da média). Isso significa que os dados possuem uma concentração maior de valores acima da média.

Skewness negativo: indica que a distribuição tem uma cauda mais longa à esquerda (mais valores à esquerda da média). Isso significa que os dados possuem uma concentração maior de valores abaixo da média.

Skewness zero: indica que a distribuição é simétrica em relação à média. Isso significa que os dados possuem a mesma concentração de valores acima e abaixo da média.

Em geral, o Skewness é uma medida útil para entender se os dados possuem uma distribuição simétrica ou assimétrica. Uma distribuição assimétrica pode afetar a precisão de alguns modelos estatísticos. Por exemplo, modelos lineares são mais precisos com distribuições simétricas.

O Kurtosis é uma medida estatística que indica a forma da distribuição de dados em relação a uma distribuição normal. Ele mede o "achatamento" ou "afilamento" da curva de distribuição em relação à curva normal. Ele pode ser positivo, negativo ou zero.

Kurtosis positivo: indica que a distribuição tem caudas mais curtas e uma "crista" mais pronunciada no meio. Isso significa que os dados possuem uma concentração maior de valores próximos à média e menos valores extremos (outliers). Esse tipo de distribuição é chamado de "leptocúrtica"

Kurtosis negativo: indica que a distribuição tem caudas mais longas e uma "crista" menos pronunciada no meio. Isso significa que os dados possuem uma concentração menor de valores próximos à média e mais valores extremos (outliers). Esse tipo de distribuição é chamado de "platocúrtica"

Kurtosis zero: indica que a distribuição é semelhante à distribuição normal.

Assim como o Skewness, o Kurtosis é uma medida útil para entender a forma da distribuição dos dados e seus possíveis efeitos em modelos estatísticos. Por exemplo, alguns modelos são mais precisos com distribuições leptocúrticas.

Percentis (ou percentis) são medidas estatísticas que indicam o percentual de valores de uma base de dados que estão abaixo de um determinado valor. Eles são usados para descrever a distribuição dos dados e para comparar as diferenças entre grupos de dados.

Os percentis são calculados dividindo o número de valores menores ou iguais a um determinado valor pelo número total de valores e multiplicando por 100. Por exemplo, o percentil 25 (ou P25) indica que 25% dos valores estão abaixo desse valor. O percentil 75 (ou P75) indica que 75% dos valores estão abaixo desse valor.

Os percentis são úteis para entender a distribuição dos dados, especialmente quando os dados possuem outliers (valores extremos) ou quando há diferenças entre grupos de dados. Por exemplo, se você tem uma base de dados de salários e quer saber qual é o salário mínimo para ser considerado no top 25% dos salários, você pode calcular o percentil 75 (P75) dos salários e ver qual é o valor correspondente. Esse valor seria o limite superior para ser considerado no top 25% dos salários.

Outra utilidade dos percentis é na definição de limites de normalidade, ou limites de valores "aceitáveis" ou "preocupantes". Por exemplo, os valores acima do percentil 97,5 (P97,5) e abaixo do percentil 2,5 (P2,5) geralmente são considerados como valores extremos ou outliers.

Os percentis também são úteis para comparar grupos de dados. Por exemplo, se você tem duas bases de dados de salários, uma para homens e outra para mulheres, você pode calcular os percentis para cada grupo e comparar os resultados para ver se há diferenças significativas entre os grupos.

É importante notar que os percentis só são úteis quando os dados estão ordenados. Além disso, é importante lembrar que os percentis não são medidas robustas, o que significa que eles podem ser afetados por outliers ou valores extremos.

A variância e a desvio padrão (ou desvio-padrão) são medidas estatísticas que indicam a dispersão dos dados em relação à média. Ambos são baseados na variação dos valores em relação à média.

Variância: é a média dos desvios ao quadrado dos valores em relação à média. Ela indica o grau de variação dos dados em relação à média. A variância é expressa em unidades quadradas (como metros quadrados ou reais quadrados)

Desvio padrão: é a raiz quadrada da variância. Ele indica o grau de variação dos dados em relação à média, mas está expressa na mesma unidade dos dados originais.

A variancia é uma medida muito utilizada para medir a variação de um conjunto de dados e ajuda a entender como os dados estão distribuidos. Porém, ela pode ser influenciada por valores extremos (outliers) e é sensível a escala, por isso, é mais comum utilizar o desvio padrão.

O desvio padrão é muito utilizado para medir a incerteza de uma amostra, e é comumente utilizado para medir a volatilidade de um ativo financeiro ou para medir a precisão de uma amostra.

O boxplot (ou "diagrama de caixa") é uma ferramenta gráfica que permite visualizar a distribuição de dados e identificar possíveis outliers (valores extremos). Ele é composto por uma caixa, que representa a maior parte dos dados (intervalo entre o primeiro quartil (Q1) e o terceiro quartil (Q3)), e por dois traços (ou "whiskers") que representam os dados fora da caixa (valores fora do intervalo entre Q1 e Q3).

A caixa do boxplot representa a faixa de valores entre o primeiro e terceiro quartil, onde a maior parte dos dados se concentra. O primeiro quartil (Q1) é o valor correspondente ao percentil 25 (P25) e o terceiro quartil (Q3) é o valor correspondente ao percentil 75 (P75). A mediana (valor central) é representada por uma linha horizontal dentro da caixa.

Os traços (ou "whiskers") representam os dados fora da caixa, ou seja, os valores que estão fora do intervalo entre o primeiro e o terceiro quartil. Eles são calculados como sendo o valor máximo menos 1.5 vezes a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1) para o lado superior, e o valor mínimo mais 1.5 vezes a diferença entre o terceiro quartil e o primeiro quartil para o lado inferior. Quaisquer valores que estejam fora desses limites são considerados outliers e são representados por pontos fora do gráfico.

O boxplot é uma ferramenta útil para comparar a distribuição de dados entre vários grupos e identificar outliers. Ele também é útil para entender a distribuição dos dados quando os dados possuem outliers, pois ele mostra a faixa de valores dentro da qual a maior parte dos dados se concentra, independentemente dos valores extremos.

# Pré-processamento

MinMaxScaler, LabelEncoder e StandardScaler são três técnicas diferentes utilizadas no pré-processamento de dados para se adequar aos algoritmos de machine learning.

MinMaxScaler: é usado para normalizar os dados, ou seja, para escalar os valores para um determinado intervalo. Ele é útil quando os dados possuem diferentes escalas ou quando os algoritmos de machine learning são sensíveis à escala dos dados. Esse Scaler é comumente utilizado para dados de entrada.

LabelEncoder: é usado para codificar variáveis categóricas em números inteiros. Ele é útil quando os algoritmos de machine learning não conseguem lidar com variáveis categóricas. Esse tipo de Scaler é utilizado para dados de entrada.

StandardScaler: é usado para padronizar os dados, ou seja, para remover a média e escalar a variação. Ele é útil quando os dados possuem diferentes escalas ou quando os algoritmos de machine learning são sensíveis à escala dos dados. Esse tipo de Scaler é utilizado para dados de entrada.

É importante notar que as técnicas de pré-processamento devem ser selecionadas de acordo com as características dos dados e do algoritmo de machine learning a ser utilizado. Alguns algoritmos podem não precisar de nenhum tipo de tratamento, enquanto outros podem precisar de mais de uma técnica.

PCA (Principal Component Analysis) é uma técnica de análise de dados utilizada para reduzir a dimensionalidade dos dados. Ele é usado para transformar um conjunto de observações de várias variáveis em um conjunto de valores de menor dimensão, chamados de componentes principais.

O objetivo do PCA é encontrar as direções de maior variância nos dados e, em seguida, projetar os dados nessas direções. Isso é feito ao encontrar as linhas de melhor ajuste (ou eixos principais) que maximizam a variância dos dados projetados.

PCA é útil para:

Reduzir a dimensionalidade dos dados, removendo redundância e ruído.
Visualizar dados de alta dimensão.
Identificar relações estruturais entre as variáveis.
Melhorar o desempenho de algoritmos de aprendizado de máquina.
É importante notar que o PCA é uma técnica linear, e não é adequado para dados não-lineares, para esses casos existe outra técnica chamada t-SNE.

A matriz de covariância é uma técnica estatística utilizada para medir a relação linear entre duas ou mais variáveis. Ela é uma matriz quadrada que contém os coeficientes de covariância entre cada par de variáveis.

Os coeficientes de covariância medem o grau de variação conjunta entre duas variáveis. Se as duas variáveis aumentarem ou diminuírem juntas, a covariância será positiva. Se uma variável aumenta enquanto a outra diminui, a covariância será negativa. Se as duas variáveis não estiverem relacionadas, a covariância será zero.

A matriz de covariância é utilizada para entender a relação linear entre as variáveis e para identificar quais variáveis estão mais relacionadas entre si. Ela é utilizada em técnicas de análise estatística como PCA (análise de componentes principais) e análise discriminante linear.

Além disso, a matriz de covariância é utilizada em métodos de otimização como o algoritmo de Gauss-Newton, e também é utilizada para inferir a probabilidade conjunta de duas variáveis aleatórias.

# Dados desbalanceados?

RandomUnderSampler e SMOTE são técnicas de reamostragem utilizadas para tratar dados desbalanceados.

RandomUnderSampler: é um método de subamostragem que remove aleatoriamente amostras da classe majoritária até que a classe minoritária tenha o mesmo número de amostras. Esse método é útil quando há muitas amostras na classe majoritária e queremos equilibrar o número de amostras entre as classes. No entanto, essa técnica pode descartar informações valiosas da classe majoritária.

SMOTE (Synthetic Minority Over-sampling Technique): é um método de sobre-amostragem que cria novas amostras sintéticas para a classe minoritária. Ele funciona selecionando amostras da classe minoritária e gerando novas amostras sintéticas entre essas amostras selecionadas. Esse método é útil quando há poucas amostras na classe minoritária e queremos aumentar o número de amostras na classe minoritária. No entanto, essa técnica pode gerar dados artificialmente que podem ser difíceis de interpretar.

Dependendo do caso de uso e do modelo escolhido, é possível utilizar uma ou mais dessas técnicas para equilibrar o conjunto de dados.

```python
# Oversampling usando SMOTE
smote = SMOTE(random_state=0)
X_train_res, y_train_res = smote.fit_resample(X_train, Y_train)

# Verificando desbalanceamento após oversampling
print("Desbalanceamento depois do oversampling: ", y_train_res.value_counts())

# Verificando desbalanceamento antes do undersampling
print("Desbalanceamento antes do undersampling: ", y_train_res.value_counts())

# Undersampling
rus = RandomUnderSampler(random_state=0)
X_train_res, y_train_res = rus.fit_resample(X_train_res, y_train_res)

# Verificando desbalanceamento após undersampling
print("Desbalanceamento depois do undersampling: ", y_train_res.value_counts())
```

Limpeza dos dados: remover valores ausentes, valores redundantes ou valores inconsistentes.

Transformação dos dados: normalizar, codificar e reduzir a dimensionalidade dos dados.

Tratamento de dados desbalanceados: utilizar técnicas de sobre-amostragem ou sub-amostragem para equilibrar o número de amostras entre as classes.

Aqui está uma lista de coisas específicas para serem feitas antes de aplicar cada um desses algoritmos:

SVC(): Esse algoritmo é sensível à escala dos dados, então é importante normalizar ou padronizar as características antes de aplicá-lo. Além disso, é importante lidar com dados categóricos se existirem, através de técnicas de codificação.

RandomForestClassifier(): Esse algoritmo é menos sensível à escala dos dados do que o SVC, mas normalizar ou padronizar as características ainda pode ajudar a melhorar o desempenho. Além disso, é importante lidar com dados categóricos se existirem.

MLPClassifier(): Esse algoritmo é sensível à escala dos dados, então é importante normalizar ou padronizar as características antes de aplicá-lo. Além disso, é importante lidar com dados categóricos se existirem, através de técnicas de codificação.

# Avaliando modelos

roc_auc_score: É uma métrica que mede a capacidade do modelo de distinguir entre as classes. Ele mede a área abaixo da curva ROC (Receiver Operating Characteristic) do modelo, sendo que quanto maior a área, melhor é a capacidade de classificação do modelo. Ela é útil para problemas de classificação binária e pode ser calculada tanto para problemas de classificação binária quanto para problemas de classificação multiclasse.

Matrix de confusão: É uma tabela que mostra as previsões do modelo comparadas com os valores reais, onde é possível verificar os verdadeiros positivos, verdadeiros negativos, falsos positivos e falsos negativos.

Sensibilidade: É uma métrica que mede a proporção de casos positivos corretamente identificados pelo modelo. Ela é calculada como verdadeiros positivos / (verdadeiros positivos + falsos negativos).

Precisão: É uma métrica que mede a proporção de previsões positivas corretas em relação ao total de previsões positivas. Ela é calculada como verdadeiros positivos / (verdadeiros positivos + falsos positivos).

Especificidade: É uma métrica que mede a proporção de casos negativos corretamente identificados pelo modelo. Ela é calculada como verdadeiros negativos / (verdadeiros negativos + falsos positivos).

Acurácia: É uma métrica que mede a proporção de previsões corretas em relação ao total de previsões feitas pelo modelo. Ela é calculada como (verdadeiros positivos + verdadeiros negativos) / total.

F1 Score: É uma métrica que combina sensibilidade e precisão, tendo em conta a proporção de verdadeiros positivos e a proporção de falsos negativos. Ela é calculada como 2 * (precisão * sensibilidade) / (precisão + sensibilidade).

Cohen_kappa_score: É uma métrica que mede a concordância entre dois rateros (ou entre o modelo e as respostas verdadeiras) além da chance. Ela é útil em problemas de classificação multiclasse, onde a acurácia simples pode ser enganosa.

O average precision score é uma métrica utilizada para avaliar a precisão de algoritmos de classificação binária. Ele é calculado como a média do precision score para diferentes níveis de recall. Em outras palavras, ele mede a relação entre o número de verdadeiros positivos e o número de falsos positivos em diferentes pontos de corte (ou thresholds) do algoritmo. Ele é especialmente útil quando a distribuição de classe é desbalanceada, pois permite avaliar a precisão do algoritmo em diferentes níveis de recall.

```python
# Calcular métricas para o modelo SVM
print("SVM")
print(classification_report(y_test, y_pred_svm))
print(confusion_matrix(y_test, y_pred_svm))
print("AUC:", roc_auc_score(y_test, y_pred_svm))
fpr, tpr, thresholds = roc_curve(y_test, y_pred_svm)
plt.plot(fpr, tpr)
plt.show()

# Calcular métricas para o modelo Random Forest
print("Random Forest")
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
print("AUC:", roc_auc_score(y_test, y_pred_rf))
fpr, tpr, thresholds = roc_curve(y_test, y_pred_rf)
plt.plot(fpr, tpr)
plt.show()

# Calcular métricas para o modelo Multilayer Perceptron
print("Multilayer Perceptron")
print(classification_report(y_test, y_pred_mlp))
print(confusion_matrix(y_test, y_pred_mlp))
```
