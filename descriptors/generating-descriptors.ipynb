## RDKIT Chem Descriptors

```python
from cheminformatics import Chem
import pandas as pd

# Caminho para o arquivo SDF de entrada
sdf_file = '../dataset/formats/df_ready_classification.sdf'

# Inicialize listas para armazenar os valores dos descritores
pesos_moleculares = []
logP_values = []

# Abra o arquivo SDF de entrada
sdf_supplier = Chem.SDMolSupplier(sdf_file)

# Itere sobre as moléculas no arquivo SDF
for mol in sdf_supplier:
    if mol is not None:
        # Calcule descritores específicos (peso molecular e logP) para cada molécula
        peso_molecular = Chem.Descriptors.MolWt(mol)
        logP = Chem.Descriptors.MolLogP(mol)
        
        pesos_moleculares.append(peso_molecular)
        logP_values.append(logP)

# Crie um DataFrame pandas com os valores dos descritores
df_descritores = pd.DataFrame({'Peso Molecular': pesos_moleculares, 'LogP': logP_values})

# Caminho de saída para o arquivo CSV
output_csv = 'descritores_moleculares.csv'

# Salve o DataFrame em um arquivo CSV
df_descritores.to_csv(output_csv, index=False)

print(f"Descritores salvos em '{output_csv}'")
```

## RDKIT ML Descriptors

```python
#! pip install rdkit
mols = [x for x in  Chem.SDMolSupplier("../dataset/formats/df_ready_classification.sdf") if x is not None]
rdkit_2d_desc = []
calc = MoleculeDescriptors.MolecularDescriptorCalculator([x[0] for x in Descriptors._descList])
header = calc.GetDescriptorNames()
for i in range(len(mols)):
    ds = calc.CalcDescriptors(mols[i])
    rdkit_2d_desc.append(ds)
df = pd.DataFrame(rdkit_2d_desc,columns=header)
df.to_csv('./generate/rdkit/dataset_ml_rdkit_descriptors.csv')
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
import pandas as pd

# Abra o arquivo SDF
sdf_file = '../dataset/formats/virtual_screening.sdf'

# Inicialize listas para armazenar os valores de todos os descritores
descritores = Descriptors._descList
nomes_descritores = [desc[0] for desc in descritores]
valores_descritores = [[] for _ in nomes_descritores]

# Calcule os descritores para cada molécula no arquivo SDF
sdf_supplier = Chem.SDMolSupplier(sdf_file)
for mol in sdf_supplier:
    if mol is not None:
        for i, desc in enumerate(descritores):
            nome_descritor = desc[0]
            valor_descritor = desc[1](mol)
            valores_descritores[i].append(valor_descritor)

# Crie um DataFrame pandas com os valores de todos os descritores
df_descritores = pd.DataFrame({nome: valores for nome, valores in zip(nomes_descritores, valores_descritores)})

# Caminho de saída para o arquivo CSV
output_csv = './generate/rdkit/virtual_screening_rdkit_chem_descriptors.csv'

# Salve o DataFrame em um arquivo CSV
df_descritores.to_csv(output_csv, index=False)
```
